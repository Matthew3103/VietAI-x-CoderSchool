```python
import pandas as pd


data = pd.read_csv('dataset_3.csv')


#Task 1: Liệt kê các sản phẩm có doanh thu lớn hơn 100.000 Euro


data['revenue'] = data['price'] * data['units_sold']
lietkesanphamdt = data[data['revenue'] > 100000]
#print(lietkesanphamdt[['title','revenue']])




#Task 2: Liệt kê các sản phẩm có units_sold nằm trong top 100, đồng thời nằm trong nhóm 10% sản phẩm có rating cao nhất


top100_units_sold = data['units_sold'].nlargest(n=100).index
top157_rating = data['rating'].nlargest(n=157).index


top_unit_and_rating = data.loc[top100_units_sold.intersection(top157_rating)]
print(top_unit_and_rating[['title', 'units_sold', 'rating']])




#Task 3: Có phải sản phẩm càng được rating nhiều thì càng bán được nhiều sản phẩm ko


correlation = data['rating'].corr(data['units_sold'])
print('Với hệ số tương quan 0.0395 giữa rating và số lượng bán, có thể nhận xét rằng rating hầu như không ảnh hưởng đến doanh số')




#Task 4: Thống kê Top 50 sản phẩm nằm trong rating từ 3.0 → 4.0 bán chạy nhất


rating_3_to_4 = data[(data['rating'] >= 3.0) & (data['rating'] <= 4.0)]
top50_units_sold = rating_3_to_4['units_sold'].nlargest(n=50).index
print(data.loc[top50_units_sold, ['title', 'units_sold', 'rating']])




#Task 5: Có bao nhiêu sản phẩm có giá bán lớn hơn 50 EURO


sosp = data[data['retail_price'] > 50]


print(sosp)



```
