```python
import pandas as pd


data =pd.read_csv('dataset_1.csv')


data['Population']=data['Population'].str.replace(',', '').astype(int)
data['Density KM2']=data['Density KM2'].str.replace(',', '').astype(int)



#Task 1: Sắp xếp và in ra 10 thành phố có dân số lớn nhất & 10 thành phố có dân số nhỏ nhất
b= data.sort_values(by=('Population'))


max10= b.tail(10)
print(max10[['City', 'Population']])


min10=b.head(10)
print(min10[['City', 'Population']])


#Task 2: In ra tên các quốc gia có tối thiểu 3 thành phố trong danh sách này


dem=data.groupby('Country').size()
dem=dem[dem>2]

print(dem)


#Task 3: In ra Top 5 quốc gia có nhiều thành phố xuất hiện trong bảng này nhất


print(dem.nlargest(n=5))




#Task 4: In ra các thành phố có dân số & diện tích đều nằm trong Top 20


ds20= data['Population'].nlargest(n=20).index
dt20= data['Area KM2'].nlargest(n=20).index
jj=data[data.index.isin(ds20) & data.index.isin(dt20)]

print('câu 4 ---------------------------', jj[['City', 'Population', 'Area KM2']]  )


#Task 5: Thống kê mật độ dân số theo quốc gia

thongkemd = data.groupby('Country')['Density KM2'].mean()
print(thongkemd)

#Task 6: Thống kê các thành phố có dân số lớn nhất của từng quốc gia (chỉ tính của những quốc gia có 2 
# thành phố xuất hiện trở lên trong bảng)

quocgia2=data.groupby('Country').size()
quocgia2=quocgia2[quocgia2>1]

dansolonnhat=data.loc[data.groupby('Country')['Population'].idxmax(), ['Country','City','Population']]
print(dansolonnhat)

```
